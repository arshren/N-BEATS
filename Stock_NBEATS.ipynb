```python
max_prediction_length=30
last_date = scaled_data.index.max()  # Assuming your DataFrame 'data' has a DatetimeIndex
future_dates = pd.date_range(last_date + pd.DateOffset(days=1), periods=max_prediction_length)
# Adjust 'periods' for the desired number of future days

#Create a DataFrame for future data:
future_data = pd.DataFrame(index=future_dates, columns=scaled_data.columns).fillna(0)

# Fill future_data with forward fill:
for column in future_data.columns:
    future_data[column] = scaled_data[column].iloc[-1]
# Fill with the last known values using forward fill
#future_data = pd.DataFrame(index=future_dates, columns=data.columns).fillna(method='ffill')
#Combine historical and future data:
scaled_data = pd.concat([scaled_data, future_data])

scaled_data.tail(5)
```

```python
def mape(y_true, y_pred):
    """
    Calculates the Mean Absolute Percentage Error (MAPE).

    Args:
        y_true: Tensor of actual values.
        y_pred: Tensor of predicted values.

    Returns:
        The MAPE value.
    """

    # Ensure tensors are on the same device
    y_true = y_true.to(y_pred.device)

    # Avoid division by zero
    mask = y_true != 0

    # Calculate absolute percentage error and average
    ape = torch.abs((y_true[mask] - y_pred[mask]) / y_true[mask])
    mape = torch.mean(ape) * 100

    return mape
```

```python
import optuna
#from pytorch_forecasting import NBeats, TimeSeriesDataSet
#from pytorch_forecasting.metrics import SMAPE
#import lightning.pytorch as pl
def objective(trial):
    # Define hyperparameter search space

    widths= trial.suggest_categorical("widths", [[32, 512], [64, 1024]])
    backcast_loss_ratio= trial.suggest_float("backcast_loss_ratio", 0.1, 1.0)
    learning_rate= trial.suggest_float("learning_rate", 1e-4, 1e-2, log=True)
    weight_decay= trial.suggest_float("weight_decay", 1e-4, 1e-2, log=True)
    num_blocks = trial.suggest_int("num_blocks_per_stack", 1, 4)
    num_stacks = trial.suggest_int("num_stacks", 1, 3)
    dropout = trial.suggest_float("dropout", 0.0, 0.5)
        # Add other hyperparameters to tune


    # Create and train N-BEATS model with trial parameters
    net = NBeats.from_dataset(
    training,
    learning_rate=learning_rate,
    log_interval=10,
    log_val_interval=1,
    weight_decay=weight_decay,
    widths=widths,
    backcast_loss_ratio=backcast_loss_ratio,
    # Pass num_blocks and num_stacks directly to NBeats, not as **kwargs
    num_blocks=[3,3],
    num_block_layers=[4,4],
    dropout=dropout
    )
    # Early stopping callback
    #early_stop_callback = EarlyStopping(monitor="val_loss", min_delta=1e-4, patience=2, verbose=False, mode="min")

    #trainer = pl.Trainer(max_epochs=5, accelerator="auto") # consider increasing max_epochs
    trainer = pl.Trainer(
    max_epochs=5,
    accelerator="auto",

    gradient_clip_val=0.01,
    #callbacks=[early_stop_callback],
    limit_train_batches=100,
    #callbacks=[early_stop_callback],
    )
    trainer.fit(
        net,
        train_dataloaders=train_dataloader,
        val_dataloaders=val_dataloader,
    )
    #Evaluate the Model
     #actuals = torch.cat([y[0] for x, y in iter(val_dataloader)])
    #predictions = net.predict(val_dataloader, mode="raw", return_x=True)

    # Extract predictions as a Tensor
    #prediction_values = predictions.output.detach().cpu()
    # .detach() to remove it from the computation graph
    # .cpu() to move it to the CPU if necessary
    val_loss= trainer.callback_metrics['val_loss'].item()
    return val_loss





# Create Optuna study and optimize
study = optuna.create_study(direction="minimize")
study.optimize(objective, n_trials=10)  # Consider increasing n_trials

# Print best hyperparameters and score
print("Best hyperparameters:", study.best_params)
print("Best SMAPE:", study.best_value)
```
